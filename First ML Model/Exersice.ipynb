## üß† Challenge 1: Fruit Guessing Upgrade

**Task:**  
Add more fruits! üçåüçáüçé  
Use features like **[weight, color_code]**

**Color Code Example:**
- Red = 0  
- Orange = 1  
- Yellow = 2 *(for banana)*  
- Purple = 3 *(for grapes)*  

---

**üîπ Hint:**
```python
features = [[150,0],[170,0],[130,1],[120,1],[110,2],[50,3]]

```

---


- Question:
    - Now predict for [100, 2] (yellow, light) ‚Äî kya lagta hai?


````python
from sklearn.tree import DecisionTreeClassifier
features = [[150,0],
[170,0],
[130,1],
[120,1],
[110,2],
[50,3]]
labels = ["apple", "apple", "orange", "orange","banana","grapes"]
clf = DecisionTreeClassifier()
clf = clf.fit(features,labels)
clf.predict([[100,2]])
````

## üéØ Challenge 3: Pass or Fail Predictor

## üè† Challenge 5: House Price Range

**Task:**  
Predict **house price category** based on **area** and **number of bedrooms**.  
*(0 = Low, 1 = Medium, 2 = High)*

---

**üîπ Hint:**
```python
features = [[800,1],[1500,2],[2500,3],[1200,2]]
labels = ["low","medium","high","medium"]
```
---
- Question:
    - Predict [2000, 3] ‚Äî kya output?

````python
from sklearn.tree import DecisionTreeClassifier
features = [
    [800,1],
    [1500,2],
    [2500,3],
    [1200,2]
]
labels = ["low","medium","high","medium"]
clf = DecisionTreeClassifier()
clf = clf.fit(features,labels)
clf.predict([[2000,3]])
````
